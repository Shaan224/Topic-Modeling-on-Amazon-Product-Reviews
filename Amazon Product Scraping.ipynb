```python
import os 
import numpy as np
import pandas as pd
import urllib.request as url
import re
import requests
from bs4 import BeautifulSoup as bs
import warnings
warnings.filterwarnings('ignore')
```

```python
def amazon_product_reviews(product_link):
    try:
        import requests
        import urllib.request as url
        from bs4 import BeautifulSoup as bs
        import pandas as pd
        import numpy as np
        import warnings
        warnings.filterwarnings('ignore')
        product_link = str(product_link)
        page = 1
        headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'}
        response = requests.get(product_link,headers=headers)
        soup = bs(response.content,'html.parser')
        review_link = 'https://www.amazon.in' + soup.find('a',class_="a-link-emphasis a-text-bold").get('href')
        
        soup3=bs(requests.get(review_link,headers=headers).content,'html.parser')
        page_number = soup3.find('div',class_='a-section a-spacing-medium').get_text()
        page_number = re.sub(",",'',page_number)
        x = page_number.split()
        max_page = int(int(x[3])/10) +1
        
        User=[]
        Ratings=[]
        Mini_Review=[]
        Date=[]
        Reviews=[]

        for page in range(1,max_page+1):

            get_review_link = review_link +'&pageNumber='+ str(page)
            soup2=bs(requests.get(get_review_link,headers=headers).content,'html.parser')
            
            for each_user in soup2.find_all('div',class_="a-section review aok-relative"):
                
                for user in each_user.find_all('div',class_="a-profile-content"):
                    User.append(user.get_text())
                for rating in each_user.find_all('span',class_="a-icon-alt"):
                    Ratings.append(rating.get_text())
                for date in each_user.find_all('span',class_="a-size-base a-color-secondary review-date"):
                    Date.append(date.get_text())    
                for mini_review in each_user.find_all('a',class_="a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"):
                    Mini_Review.append(mini_review.get_text())   
                for review in each_user.find_all('div',class_="a-row a-spacing-small review-data"):
                    Reviews.append(review.get_text())

            
        
        

        Data = list(zip(User,Ratings,Mini_Review,Date,Reviews))
        Review_data = pd.DataFrame(Data,columns=['User Name','Ratings','Mini Review','Date','Reviews'])    
        return Review_data
    except:
        pass
```

```python
Review_data = amazon_product_reviews('https://www.amazon.in/Mivi-ThunderBeats-Wireless-Bluetooth-Earphones/dp/B07VNL73FF/ref=sr_1_1?_encoding=UTF8&pf_rd_i=desktop&pf_rd_m=A1VBAL9TL5WCBF&pf_rd_p=72d48896-4d71-41bb-878b-961db227d942&pf_rd_r=E8C74Z3X67AS097GMW9A&pf_rd_t=36701&qid=1580288840&s=electronics&smid=A14CZOWI0VEHLG&sr=1-1')
```
